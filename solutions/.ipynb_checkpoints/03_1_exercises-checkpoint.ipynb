```python
import numpy as np
from matplotlib import pyplot as plt
d = np.load('exampleCAtracenpy.sec')
time_axis = np.linspace(0, len(d)/20, len(d))
#print(d, time_axis)
#print(len(d), len(time_axis))
pl = plt.figure(figsize=(11, 6))
plt.plot(time_axis, d)
plt.xlabel('Time, s')
plt.xlim(0, len(d)/20)
plt.show()
```

```python
plt.plot(time_axis[1000:2000], d[1000:2000])
plt.xlabel('Time, s')
plt.xlim(50, 100)
plt.show()
```

```python
import math
def gaus_kernel(sigma, size = 100):
    t = np.linspace(-size//20, size//20, size*2)
    sg = sigma/1000
    y = (1 / (sg * (math.sqrt(2 * np.pi)))) * (np.exp ((-t ** 2)/(2 * (sg**2))))
    plt.plot(t, y)
    return y
 
gaus_kernel(250)
#plt.plot(t, gaus_kernel(250))
```

```python
sigma = 250
after_conv = np.convolve(d[1000:2000], gaus_kernel(sigma), 'same')

plt.plot(time_axis[1000:2000], after_conv)
plt.title('After convolution, \u03C3 = {} ms'.format(sigma))
plt.show()
```

```python
def sin_kernel(size = 100):
    t = np.linspace(-size//20, size//20, size*2)
    sin = [-(math.sin(2 * math.pi * t1)) for t1 in t]
 #   plt.plot(t, sin)
    return sin

#sin_kernel()
#plt.plot(sin_kernel())
#plt.show()
```

```python
def new_kernel(sigma):
    gaus = gaus_kernel(sigma)
    sin = np.array(sin_kernel())
    new = gaus * sin
    return new
#new_kernel(250)
plt.plot(new_kernel(250))
```

```python
sigma = 250
aft_conv = np.convolve(d[1000:2000], new_kernel(sigma), 'same')

fig, ax = plt.subplots(3, 1, figsize = (10, 10))
ax[0].plot(time_axis[1000:2000], d[1000:2000])
ax[0].set_title('Before convolution')
ax[1].plot(time_axis[1000:2000], after_conv)
ax[1].set_title('After convolution with gaussian,\n \u03C3 = {} ms'.format(sigma))
ax[2].plot(time_axis[1000:2000], aft_conv)
ax[2].set_title('After convolution with \n gaussian * sin')
plt.tight_layout()


plt.show()
# Gaussian * sin kernel look very much like a derivative of a Gaussian and 
# convolution of a fucntion with a derivative is a derivative of a convolution
# To generate a second derivative, I can approximate second derivative of a gaussian
```

```python
def kernel_for_sec_dev(sigma, size= 100):
    first_kernel = gaus_kernel(sigma)
    
    second_kernel = new_kernel(sigma)

    t = np.linspace(-size//20, size//20, size * 2)
    function = [-(math.cos(2 * math.pi * t2)) for t2 in t] 

    new_kern = first_kernel*function
 
    plt.plot(new_kern)

    return new_kern
kernel_for_sec_dev(250)
```

```python
sec_der = np.convolve(d[1000:2000], kernel_for_sec_dev(250), 'same')

fig1, ax = plt.subplots(3, 1, figsize = (13, 9))
ax[0].plot(time_axis[1000:2000], after_conv)
ax[0].set_title('After convolution with gaussian,\n \u03C3 = {} ms'.format(sigma))
ax[1].plot(time_axis[1000:2000], aft_conv)
ax[1].set_title('After convolution with \n gaussian * sin \n approximation of 1st derivative')
ax[2].plot(time_axis[1000:2000], sec_der)
ax[2].set_title('After convolution with \n gaussian * cos \n approxiamtion of 2nd derivative')
plt.tight_layout()
```
